```python
reviewText=list(data["Review Text"].replace(np.nan,"NA"))
```

```python
from nltk.tokenize import sent_tokenize
sentText=[sent_tokenize(text) for text in reviewText]
```

```python
sentText[:5]
```

```python
print(len(sentText))
```

```python
from nltk.tokenize import word_tokenize
import string
processed_sents=[[word_tokenize(item.lower()) for item in text] for text in sentText]
stopwords_set=set(stopwords.words("english"))
i=0
for text in processed_sents :
    j=0
    for sentence in text: 
        processed_list=[]
        for word in sentence :
            if word not in stopwords_set and word not in string.punctuation :
                processed_list.append(word)
        processed_sents[i][j]=processed_list
        j+=1
    i+=1
```

```python
print(processed_sents[:5])
```

```python
from nltk import pos_tag
pos_tags=[]
for text in processed_sents :
    empty_list=[]
    for item in text :
        empty_list.append(pos_tag(item))
    pos_tags.append(empty_list)
```

```python
pos_tags[:5]
```

```python
def extractor(processed_sents,tag1,tag2) :
    ret_list=[]
    i=0
    for text in processed_sents :
        text_info=[]
        for sentence in text :
            sent_info=[]
            for item in sentence :
                word=item[0]
                tag=item[1]
                if tag==tag1 or tag==tag2 :
                    sent_info.append(item)
            if len(sent_info) != 0 :
                text_info.append(sent_info)
        ret_list.append(text_info)
        if i%100 ==0 :
            print(text_info)
        i+=1
    return ret_list
                
```

```python
noun_adj_list=extractor(pos_tags,"NN","JJ")
```

```python
def penn_to_wn(tag):
    """
    Convert between the PennTreebank tags to simple Wordnet tags
    """
    if tag.startswith('J'):
        return wn.ADJ
    elif tag.startswith('N'):
        return wn.NOUN
    elif tag.startswith('R'):
        return wn.ADV
    elif tag.startswith('V'):
        return wn.VERB
    return None
```

```python
from nltk.corpus import wordnet as wn
from nltk.corpus import sentiwordnet as swn
sent_scores=[]
for text in noun_adj_list :
    sent_score=[]
    for sentence in text :
        sentiment=0
        for element in sentence: 
            syn=wn.synsets(element[0],penn_to_wn(element[1]))
            if len(syn) >0 :
                swn_syn=swn.senti_synset(syn[0].name())
                sentiment+=swn_syn.pos_score()-swn_syn.neg_score()
        sent_score.append(sentiment)
    sent_scores.append(sent_score)
```
